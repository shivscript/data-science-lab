## 1. Effective Data Visualizations

Data visualization helps us to understand the patterns and insights hidden in the data. Below are examples of different types of visualizations:

### Line Charts
Used to visualize trends over time.

#### Example: Sales over a year


````python
# Import the necessary libraries
import matplotlib.pyplot as plt  # For creating visualizations

# Define the sample data
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']  # List of months
sales = [200, 220, 250, 270, 300, 310, 400, 420, 410, 390, 370, 350]  # Corresponding sales data for each month

# Create a new figure for the plot with a specified size
plt.figure(figsize=(10, 6))  # The plot will be 10 inches wide and 6 inches tall

# Plot the data on a line chart
plt.plot(
    months,             # x-axis data (months)
    sales,              # y-axis data (sales)
    marker='o',         # Adds circular markers at each data point
    color='b',          # Line color (blue)
    linestyle='-',      # Line style (solid line)
    label='Sales'       # Label for the legend
)

# Add a title to the plot
plt.title('Monthly Sales')

# Label the x-axis
plt.xlabel('Months')

# Label the y-axis
plt.ylabel('Sales')

# Add a legend to the plot
plt.legend()

# Add a grid to the plot for better readability
plt.grid(True)

# Display the plot
plt.show()
````


### Bar Charts
Used to compare quantities.

#### Example: Sales by product category


````python

import matplotlib.pyplot as plt

# Sample data
categories = ['Electronics', 'Fashion', 'Home', 'Books', 'Beauty']
sales = [500, 700, 400, 300, 450]

# Plot bar chart
plt.figure(figsize=(10, 6))
plt.bar(categories, sales, color='skyblue')
plt.title('Sales by Product Category')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.show()
````


### Heat Maps
Used to visualize correlations or density.

#### Example: Correlation matrix


````python
# Import necessary libraries
import seaborn as sns       # For creating statistical data visualizations
import numpy as np          # For numerical operations and creating random data
import pandas as pd         # For data manipulation and analysis

# Generate sample data
data = np.random.rand(10, 4)  # Creates a 10x4 matrix of random numbers between 0 and 1
columns = ['Feature1', 'Feature2', 'Feature3', 'Feature4']  # Column names for the data
df = pd.DataFrame(data, columns=columns)  # Create a DataFrame from the random data with the defined column names

# Calculate the correlation matrix
corr = df.corr()  # Computes pairwise correlation of columns in the DataFrame

# Plot the heatmap for the correlation matrix
plt.figure(figsize=(8, 6))  # Set the figure size to 8 inches wide and 6 inches tall
sns.heatmap(
    corr,                # Data to be visualized (correlation matrix)
    annot=True,          # Annotate the cells with the correlation values
    cmap='coolwarm',     # Use the 'coolwarm' colormap for a visual gradient
    fmt='.2f'            # Format the annotation to show numbers with 2 decimal places
)
plt.title('Correlation Heatmap')  # Add a title to the heatmap
plt.show()  # Display the heatmap
````

## 2. Matrix Factorization and Gradient Descent

### Matrix factorization
Matrix factorization is a mathematical technique that breaks down a large matrix into smaller matrices that, when multiplied, approximate the original. It is widely used in recommendation systems, data compression, and dimensionality reduction.

Matrix factorization is a technique widely used in recommendation systems to predict user-item relationships. 

#### Real-World Example
Netflix Recommendation System: Netflix predicts a user's preference for movies they haven't rated yet. If a user has rated "Inception" and "The Dark Knight" highly, matrix factorization can help predict that the user might also like other movies by Christopher Nolan, such as "Interstellar."

### Gradient descent
Gradient descent is an optimization algorithm used to minimize a loss function by iteratively adjusting parameters. It is widely used in machine learning, deep learning, and optimization problems.

#### Real-World Example
Gradient descent is like perfecting a pancake recipe based on feedback. You start with a guess for the amounts of ingredients (flour, milk, eggs, sugar), make pancakes, and get feedback (e.g., "too sweet, too thick"). You then adjust the ingredients slightly, reducing sugar or flour, and try again. Each adjustment reduces the error (how far the pancake is from perfect). Over time, with small, repeated changes, you get closer to the ideal recipe. Similarly, gradient descent adjusts parameters in a model step-by-step to minimize errors and improve predictions.

### **Types of Gradient Descent**

1. **Batch Gradient Descent**  
   - Uses the entire dataset to compute the gradient at each step.  
   - Pros: More stable updates, less noise in gradient calculations.  
   - Cons: Computationally expensive for large datasets.  

2. **Stochastic Gradient Descent (SGD)**  
   - Updates parameters for each data point individually.  
   - Pros: Faster updates, suitable for online learning.  
   - Cons: Noisy updates, which can lead to fluctuating convergence.  

3. **Mini-batch Gradient Descent**  
   - Combines Batch Gradient Descent and SGD by using a small subset (mini-batch) of the data for each update.  
   - Pros: Balances speed and stability, reduces noise compared to SGD.  
   - Cons: Still computationally expensive for very large datasets.  


## Matrix Factorization with Gradient Descent Example in Python

### Problem:
We have a small dataset where users rate movies. Some ratings are missing, and we want to predict them.

**User-Movie Rating Matrix**:

|        | Movie 1 | Movie 2 | Movie 3 |  
|--------|---------|---------|---------|  
| User 1 |    5    |    3    |    ?    |  
| User 2 |    4    |    ?    |    2    |  
| User 3 |    ?    |    1    |    3    |  

The goal is to predict the missing ratings (denoted as "?").

---

````python
import numpy as np

# Small user-movie rating matrix (0 indicates missing ratings)
ratings = np.array([  # Create a 3x3 matrix of ratings where 0 represents missing ratings
    [5, 3, 0],  # User 1's ratings for Movie 1 and Movie 2, missing for Movie 3
    [4, 0, 2],  # User 2's ratings for Movie 1 and Movie 3, missing for Movie 2
    [0, 1, 3]   # User 3's ratings for Movie 2 and Movie 3, missing for Movie 1
])

# Initialize user (P) and movie (Q) matrices with random values
num_users, num_movies = ratings.shape  # Get the dimensions of the ratings matrix (3 users, 3 movies)
num_features = 2  # Number of latent features to be learned (used to represent users and movies)
P = np.random.rand(num_users, num_features)  # Randomly initialize the user-feature matrix (P)
Q = np.random.rand(num_movies, num_features)  # Randomly initialize the movie-feature matrix (Q)

# Transpose Q for easier calculations
Q = Q.T  # Transpose the Q matrix to make matrix multiplication easier

# Training parameters
steps = 2000  # Number of iterations (epochs) for the gradient descent
learning_rate = 0.01  # The step size for updating the features in each iteration
regularization = 0.02  # Regularization term to avoid overfitting

# Matrix factorization function
def matrix_factorization(R, P, Q, steps, learning_rate, reg):
    for step in range(steps):  # Loop through the number of iterations
        for i in range(len(R)):  # Loop through each user (row in the ratings matrix)
            for j in range(len(R[i])):  # Loop through each movie (column in the ratings matrix)
                if R[i][j] > 0:  # Only update for known ratings (non-zero entries in the ratings matrix)
                    # Error between the actual and predicted ratings
                    error = R[i][j] - np.dot(P[i, :], Q[:, j])  # Compute the prediction error
                    # Update user and movie features using gradient descent
                    P[i, :] += learning_rate * (error * Q[:, j] - reg * P[i, :])  # Update user feature vector (P)
                    Q[:, j] += learning_rate * (error * P[i, :] - reg * Q[:, j])  # Update movie feature vector (Q)
    return P, Q  # Return the updated user and movie matrices

# Perform matrix factorization
P, Q = matrix_factorization(ratings, P, Q, steps, learning_rate, regularization)  # Train the matrices P and Q

# Reconstruct the full rating matrix (predicted ratings)
predicted_ratings = np.dot(P, Q)  # Multiply the user matrix (P) and movie matrix (Q) to get the predicted ratings

print("Original Ratings Matrix:")  # Print the original ratings matrix
print(ratings)  # Display the original matrix

print("\nPredicted Ratings Matrix:")  # Print the predicted ratings matrix
print(np.round(predicted_ratings, 2))  # Round the predicted ratings to 2 decimal places and display
````

## Simple Analogy
### Matrix Factorization:
Breaking a cake recipe into ingredients (flour, sugar, etc.) to recreate similar cakes.

### Gradient Descent:
Tasting the cake, realizing its too sweet, and adjusting the recipe little by little until its perfect.
