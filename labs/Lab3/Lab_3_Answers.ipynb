## 1. Line Chart Practice Question
**Question:**
You are given a dataset of monthly rainfall (in mm) for your city over the past year. Create a line chart that shows the rainfall trend for the year. Label the axes and provide a title.

**Dataset**
- January: 45mm
- February: 38mm
- March: 60mm
- April: 70mm
- May: 75mm
- June: 80mm
- July: 90mm
- August: 85mm
- September: 72mm
- October: 50mm
- November: 40mm
- December: 30mm

**Instructions:**
- Label the x-axis as "Month" and the y-axis as "Rainfall (mm)".
- Title the chart as "Monthly Rainfall Trend Created by [Your Name]".

---

````python

import matplotlib.pyplot as plt

# Data
months = ["January", "February", "March", "April", "May", "June", "July", 
          "August", "September", "October", "November", "December"]
rainfall = [45, 38, 60, 70, 75, 80, 90, 85, 72, 50, 40, 30]

# Plot
plt.figure(figsize=(10, 5))
plt.plot(months, rainfall, marker='o', color='blue', label="Rainfall (mm)")
plt.xlabel("Month")
plt.ylabel("Rainfall (mm)")
plt.title("Monthly Rainfall Trend Created by Shiva Shrestha")
plt.grid(True)
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
````

## 2. Bar Chart Practice Question
**Question:**
You are given data showing the number of books read by students in your class during the past month. Create a bar chart to compare the number of books read by each student.

**Dataset**
- Student A: 5 books
- Student B: 7 books
- [Your Name]: 3 books
- Student D: 8 books
- Student E: 6 books

**Instructions:**
- Label the x-axis with the student names and the y-axis with the number of books.
- Title the chart as "Books Read by Students".

---

````python

# Data
students = ["Student A", "Student B", "Shiva Shrestha", "Student D", "Student E"]
books_read = [5, 7, 3, 8, 6]

# Plot
plt.figure(figsize=(8, 5))
plt.bar(students, books_read, color='skyblue')
plt.xlabel("Students")
plt.ylabel("Number of Books")
plt.title("Books Read by Students")
plt.show()
````

## 3. Scatter Plot Practice Question

**Question:**
You are given data showing the relationship between study hours and grade scores for 5 students. Create a scatter plot to visualize how study time and grades are related.

**Dataset**
| Student     | Study Hours | Grade Score |
|-------------|-------------|-------------|
| Student 1   | 5           | 75          |
| Student 2   | 3           | 65          |
| Student 3   | 8           | 85          |
| [Your Name] | 2           | 60          |
| Student 5   | 6           | 80          |

**Instructions:**
- Create a scatter plot with "Study Hours" on the x-axis and "Grade Score" on the y-axis.
- Label the axes and add a title to the plot.
- Optionally, you can add a regression line to visualize the trend.

---

````python

import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Data
students = ['Student 1', 'Student 2', 'Student 3', 'Shiv', 'Student 5']
study_hours = [5, 3, 8, 2, 6]
grade_scores = [75, 65, 85, 60, 80]

# Create the scatter plot
plt.figure(figsize=(8,6))
plt.scatter(study_hours, grade_scores, color='blue')

# Label the axes
plt.xlabel('Study Hours')
plt.ylabel('Grade Score')

# Add a title
plt.title('Study Hours vs Grade Scores')

# Optionally, add a regression line
sns.regplot(x=study_hours, y=grade_scores, scatter=False, color='red')

# Show the plot
plt.show()
````

## 4. Box Plot Practice Question
**Question:**
You have the exam scores of 10 students in your class. Create a box plot to show the distribution of exam scores.

**Dataset**
- 50, 65, 70, 85, 60, 75, 80, 90, 55, 70.

**Instructions:**
- Create a box plot to show the spread of the exam scores.
- Label the x-axis as "Exam Scores" and provide a title such as "Distribution of Exam Scores Created by [Your Name]".

---

````python
import matplotlib.pyplot as plt

# Data: A list of exam scores
scores = [50, 65, 70, 85, 60, 75, 80, 90, 55, 70]

# Create a new figure with specified size
plt.figure(figsize=(6, 4))  # Width: 6 inches, Height: 4 inches

# Create a box plot for the given scores
plt.boxplot(
    scores,                  # Data to be visualized
    patch_artist=True,       # Enables color filling inside the box
    boxprops=dict(facecolor="lightblue")  # Sets the box fill color to light blue
)

# Customize the x-axis ticks
plt.xticks(
    [1],                    # Specifies the x-axis position for the single tick (index 1 for 1st box plot)
    ["Exam Scores"]         # Sets the label for the tick as "Exam Scores"
)

# Add a title to the plot
plt.title("Distribution of Exam Scores Created by Shiva Shrestha")

# Display the plot
plt.show()
````

## 6. Heatmap Visualization Practice Question

**Question:**
You are given data showing the scores of students across different subjects. Create a heatmap to visualize the relationship between students and their scores in each subject.

**Dataset:**

| Student   | Math | English | Science | History | Art |
|-----------|------|---------|---------|---------|-----|
| Student 1 | 90   | 80      | 85      | 75      | 88  |
| Student 2 | 78   | 85      | 92      | 70      | 80  |
| Student 3 | 95   | 92      | 89      | 80      | 90  |
| [Your Name] | 80   | 78      | 85      | 75      | 82  |
| Student 5 | 88   | 80      | 90      | 85      | 91  |

**Instructions:**
- Create a heatmap with students on the y-axis and subjects on the x-axis.
- Use the `seaborn.heatmap()` function to visualize the scores.
- Customize the color scale to represent score intensity.
- Label the axes and add a title to the plot.

````python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Dataset
data = {
    'Math': [90, 78, 95, 80, 88],
    'English': [80, 85, 92, 78, 80],
    'Science': [85, 92, 89, 85, 90],
    'History': [75, 70, 80, 75, 85],
    'Art': [88, 80, 90, 82, 91]
}

# Create a DataFrame
df = pd.DataFrame(data, index=['Student 1', 'Student 2', 'Student 3', 'Shiv', 'Student 5'])

# Create the heatmap
plt.figure(figsize=(8,6))
sns.heatmap(df, annot=True, cmap='coolwarm', cbar=True, linewidths=0.5)

# Add labels and title
plt.title('Student Scores Across Subjects')
plt.xlabel('Subjects')
plt.ylabel('Students')

# Show the plot
plt.show()
````
