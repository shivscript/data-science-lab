## Hypothesis Testing
Hypothesis testing is a fundamental statistical method for making decisions about a population based on sample data. It involves evaluating whether a specific assumption (the null hypothesis) is supported by the data or should be rejected in favor of an alternative hypothesis.

---

### Key Concepts:
1. **Null Hypothesis (H₀):**  
   The default assumption that there is no effect or difference. Example: "The means of two groups are equal."

2. **Alternative Hypothesis (H₁):**  
   The claim we want to test. Example: "The means of two groups are not equal."

3. **Significance Level (α):**  
   A threshold probability (commonly 0.05) to determine whether to reject H₀. It represents the risk of rejecting H₀ when it is true.

4. **Test Statistic:**  
   A value calculated from the sample data, used to assess how far the data is from what H₀ predicts.

5. **P-value:**  
   The probability of observing the test statistic (or more extreme values) under the assumption that H₀ is true. A small p-value (less than α) indicates strong evidence against H₀.

---

### Hypothesis Testing Steps:
1. **Formulate Hypotheses:**  
   Define H₀ and H₁ clearly. For instance:  
   - H₀: μ₁ = μ₂ (means are equal).  
   - H₁: μ₁ ≠ μ₂ (means are different).

2. **Select a Test:**  
   Choose an appropriate statistical test based on the data and type of analysis (e.g., t-test, ANOVA, chi-square test).

3. **Set Significance Level:**  
   Determine the α value (e.g., 0.05).

4. **Calculate the Test Statistic and P-value:**  
   Use the data to compute these values.

5. **Make a Decision:**  
   - If p-value < α, reject H₀ (evidence supports H₁).  
   - If p-value ≥ α, fail to reject H₀ (insufficient evidence to support H₁).

---

### Example:
Suppose we want to compare the average scores of two groups of students.

- **H₀:** The average scores are the same for both groups.  
- **H₁:** The average scores are different.  

We perform a **t-test** to calculate the t-statistic and p-value. If the p-value is less than 0.05, we conclude that the scores differ significantly.

---

This framework helps researchers and analysts draw meaningful conclusions from data while controlling the risk of incorrect decisions.

## Common Statistical Tests in Hypothesis Testing

### 1. **T-Test**:
- **Purpose:**  
  Compares the means of two groups to determine if they are significantly different.
- **Types:**  
  - **Independent T-Test:** Used when the groups are independent (e.g., comparing test scores of students from two different classes).  
  - **Paired T-Test:** Used when the groups are related (e.g., before-and-after measurements on the same individuals).  
- **Example:**  
  "Is the average height of males different from females in a sample?"

---

### 2. **Z-Test**:
- **Purpose:**  
  Tests whether a sample mean differs significantly from a known population mean (or compares the means of two large samples).
- **When to Use:**  
  - When the sample size is large (n > 30) or the population standard deviation is known.
- **Example:**  
  "Is the average IQ of students in a school significantly different from the national average of 100?"
- **Comparison with T-Test:**  
  - A t-test is used when the sample size is small or the population standard deviation is unknown, while a z-test requires a known standard deviation and works well with larger sample sizes.

---

### 3. **ANOVA (Analysis of Variance)**:
- **Purpose:**  
  Tests for differences among the means of three or more groups.
- **How it Works:**  
  ANOVA analyzes the variance within each group and between groups to determine if the group means differ significantly.
- **Example:**  
  "Do students from different departments (e.g., Engineering, Arts, Science) have different average GPAs?"
- **Note:**  
  If ANOVA detects a difference, follow-up tests (like Tukey's test) can identify which groups differ.

---

### 4. **Chi-Square Test**:
- **Purpose:**  
  Examines the relationship between two categorical variables.
- **Types:**  
  - **Chi-Square Test of Independence:** Determines if two variables are independent (e.g., gender and preference for a product).  
  - **Chi-Square Goodness-of-Fit Test:** Tests if observed data fits a specific distribution (e.g., dice rolls being fair).
- **Example:**  
  "Is there an association between smoking (yes/no) and lung disease (present/absent)?"

---

## 1. Hypothesis Testing
### Introduction
Hypothesis testing is a statistical method used to make decisions based on data. The process involves:
1. Defining null (H0) and alternative (H1) hypotheses.
2. Choosing a significance level (e.g., 0.05).
3. Calculating a test statistic and p-value.
4. Making a decision to accept or reject H0.


### Steps:
1. Import necessary libraries.
2. Load a dataset or generate synthetic data.
3. Perform hypothesis testing using `scipy.stats`.
4. Interpret the results.


```python
# Import necessary libraries
import numpy as np  # Importing numpy for generating synthetic data
from scipy.stats import ttest_ind  # Importing t-test function from scipy.stats

# Generate synthetic data for two groups
np.random.seed(42)  # Setting the seed for reproducibility of results
group1 = np.random.normal(50, 10, 30)  # Generate 30 samples from a normal distribution with mean 50 and standard deviation 10 for group 1
group2 = np.random.normal(55, 10, 30)  # Generate 30 samples from a normal distribution with mean 55 and standard deviation 10 for group 2

# Perform an independent t-test
t_stat, p_value = ttest_ind(group1, group2)  # Perform a t-test comparing the means of group1 and group2

# Display results
print("T-statistic:", t_stat)  # Print the calculated t-statistic
print("P-value:", p_value)  # Print the p-value obtained from the t-test

# Decision based on significance level
alpha = 0.05  # Set the significance level (alpha) for hypothesis testing
if p_value < alpha:  # If the p-value is less than the significance level
    print("Reject the null hypothesis (H0): The means are significantly different.")  # Reject the null hypothesis, the means are significantly different
else:  # If the p-value is greater than or equal to the significance level
    print("Fail to reject the null hypothesis (H0): The means are not significantly different.")  # Fail to reject the null hypothesis, the means are not significantly different
```

## 2. Central Limit Theorem (CLT)
The **Central Limit Theorem (CLT)** is a fundamental concept in statistics that describes the behavior of sample means when drawn from a population:

### 1. **Definition:**
The CLT states that, regardless of the population's original distribution, the distribution of the sample mean will approximate a normal distribution as the sample size increases, provided the samples are independent and identically distributed.

### 2. **Key Points:**
- The larger the sample size (\\(n\\)), the closer the sample mean's distribution is to a normal distribution.
- The theorem applies to various population distributions (e.g., uniform, skewed, or bimodal).
- The mean of the sampling distribution equals the population mean (\\(\\mu\\)).
- The standard deviation of the sampling distribution is called the **standard error** and is calculated as:
  $$
  \sigma_{\bar{x}} = \frac{\sigma}{\sqrt{n}}
  $$
  where,
  $$
  \sigma\
  $$ is the population standard deviation.

### 3. **Why is CLT Important?**
- It allows statisticians to make inferences about population parameters using sample statistics.
- It underpins many statistical methods, such as hypothesis testing and confidence intervals.

### 4. **Practical Example:**
Suppose we are measuring the average height of individuals in a city. Even if the height distribution is skewed, by taking multiple random samples (each with a sufficient size, e.g., \\(n > 30\\)), the average height of these samples will form a bell-shaped curve, approximating a normal distribution.


```python
import matplotlib.pyplot as plt

# Generate a non-normal population dataset
np.random.seed(42)  # Setting the seed for reproducibility of results
population = np.random.exponential(scale=2, size=10000)  # Generating a population with an exponential distribution (mean=2) and 10,000 samples

# Parameters
sample_size = 30  # Defining the size of each sample
num_samples = 1000  # Number of samples to draw for calculating sample means

# Collect sample means
# We generate 1000 samples, each of size 30, and calculate the mean of each sample
sample_means = [np.mean(np.random.choice(population, sample_size)) for _ in range(num_samples)]

# Plot results
plt.figure(figsize=(10, 6))  # Create a figure for the plot with a specified size
plt.hist(sample_means, bins=30, color='skyblue', edgecolor='black', alpha=0.7)  # Plot a histogram of the sample means
plt.title("Sampling Distribution of the Sample Mean")  # Set the title of the plot
plt.xlabel("Sample Mean")  # Label the x-axis
plt.ylabel("Frequency")  # Label the y-axis
plt.axvline(np.mean(population), color='red', linestyle='dashed', linewidth=2, label="Population Mean")  # Add a vertical line for the population mean
plt.legend()  # Display the legend for the plot
plt.show()  # Show the plot
```
