## 1. NumPy: Numerical Python

NumPy is a library used for numerical computations. It provides:
- Support for multi-dimensional arrays.
- Functions for mathematical and statistical operations.

### Basic Operations
Below are examples to get started with NumPy arrays and operations.


````python

import numpy as np

# Creating a NumPy array
arr = np.array([1, 2, 3, 4, 5])
print("Array:", arr)

# Array operations
print("Array multiplied by 2:", arr * 2)
print("Array mean:", np.mean(arr))
print("Array standard deviation:", np.std(arr))
````


### Practice Questions:
1. Create a NumPy array with numbers from 1 to 10 and calculate the sum.
2. Create a 3x3 matrix with random numbers and find its transpose.
3. Generate an array of 50 evenly spaced numbers between 0 and 5.


````python
# Create a NumPy array with numbers from 1 to 10 and calculate the sum.

import numpy as np

array_1_to_10 = np.arange(1, 11)  # NumPy array with numbers from 1 to 10
sum_array = np.sum(array_1_to_10)  # Calculating the sum

print("Array:", array_1_to_10)
print("Sum:", sum_array)
````

````python
# Create a 3x3 matrix with random numbers and find its transpose.

import numpy as np

random_matrix = np.random.rand(3, 3)  # Creating a 3x3 matrix with random numbers
transpose_matrix = random_matrix.T  # Transposing the matrix

print("Original Matrix:\n", random_matrix)
print("Transposed Matrix:\n", transpose_matrix)
````

````python
# Generate an array of 50 evenly spaced numbers between 0 and 5:

import numpy as np

evenly_spaced_array = np.linspace(0, 5, 50)  # Generating 50 evenly spaced numbers

print("Evenly Spaced Array:\n", evenly_spaced_array)
````


## 2. Pandas: Data Manipulation

Pandas is a library for handling structured data. It provides:
- DataFrame and Series structures for data handling.
- Functions for data cleaning, filtering, and grouping.

### Basic Operations
Below are examples to get started with Pandas.


````python

import pandas as pd

# Creating a DataFrame
data = {'Name': ['Alice', 'Bob', 'Charlie'], 'Age': [25, 30, 35], 'Score': [85, 90, 95]}
df = pd.DataFrame(data)
print("DataFrame:\n", df)

# Basic operations
print("Mean age:", df['Age'].mean())
print("Data sorted by Score:\n", df.sort_values(by='Score', ascending=False))
````

````python
import pandas as pd

# Sample dataset
data = [12, 15, 14, 10, 18, 20, 15, 14]

# Convert to a pandas DataFrame
df = pd.DataFrame(data, columns=['Values'])

# Descriptive statistics
mean = df['Values'].mean()
median = df['Values'].median()
std_dev = df['Values'].std()
variance = df['Values'].var()

print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Standard Deviation: {std_dev}")
print(f"Variance: {variance}")
````


### Practice Questions:
1. Create a DataFrame for 5 students with their names, marks, and grades. Find the average marks.
2. Load a CSV file into a DataFrame and filter rows where marks are greater than 70.
3. Group the DataFrame by a column and calculate the sum of another column for each group.


````python
# Create a DataFrame for 5 students with their names, marks, and grades. Find the average marks:

import pandas as pd

# Creating the DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Marks': [85, 78, 92, 65, 88],
    'Grade': ['A', 'B', 'A', 'C', 'A']
}
df = pd.DataFrame(data)

# Calculating the average marks
average_marks = df['Marks'].mean()

print("DataFrame:\n", df)
print("Average Marks:", average_marks)
````

````python
# Create students.csv file for below analysis

import csv

# Create a CSV file named 'students.csv'
with open('students.csv', 'w', newline='') as csvfile:
    fieldnames = ['Name', 'Marks', 'Grade']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

    # Write the header row
    writer.writeheader()

    # Write sample data
    writer.writerow({'Name': 'Alice', 'Marks': 85, 'Grade': 'A'})
    writer.writerow({'Name': 'Bob', 'Marks': 72, 'Grade': 'B'})
    writer.writerow({'Name': 'Charlie', 'Marks': 90, 'Grade': 'A+'})
    writer.writerow({'Name': 'David', 'Marks': 68, 'Grade': 'C'})
    writer.writerow({'Name': 'John', 'Marks': 90, 'Grade': 'A+'})
````

````python
# Load a CSV file into a DataFrame and filter rows where marks are greater than 70:

import pandas as pd

# Assuming a CSV file named 'students.csv' with columns 'Name', 'Marks', and 'Grade'
df = pd.read_csv('students.csv')

# Filtering rows where marks are greater than 70
filtered_df = df[df['Marks'] > 70]

print("Filtered DataFrame:\n", filtered_df.to_string(index=False)) # disable showing index value
````

````python
# Group the DataFrame by a column and calculate the sum of another column for each group:

# Grouping by 'Grade' and calculating the sum of 'Marks'
grouped_sum = df.groupby('Grade')['Marks'].sum()

print("Sum of Marks by Grade:\n", grouped_sum)

````


## 3. Data Visualization: Matplotlib and Seaborn

Matplotlib and Seaborn are libraries for visualizing data through graphs and plots.

### Basic Operations
Below are examples to create visualizations.


````python

import matplotlib.pyplot as plt
import seaborn as sns

# Sample data
data = [12, 15, 14, 10, 18, 20, 15, 14]

# Line plot using Matplotlib
plt.plot(data, marker='o')
plt.title("Line Plot")
plt.xlabel("Index")
plt.ylabel("Value")
plt.show()

# Boxplot using Seaborn
sns.boxplot(data=data)
plt.title("Boxplot")
plt.show()
````


### Practice Questions:
1. Create a bar chart for sales data of 5 products using Matplotlib.
2. Generate a histogram for random data of size 100 using Seaborn.
3. Create a scatter plot with a regression line using Seaborn.


````python
# Create a bar chart for sales data of 5 products using Matplotlib:

import matplotlib.pyplot as plt

# Sales data
products = ['Product A', 'Product B', 'Product C', 'Product D', 'Product E']
sales = [120, 150, 90, 200, 170]

# Creating the bar chart
plt.bar(products, sales, color='skyblue')
plt.title('Sales Data')
plt.xlabel('Products')
plt.ylabel('Sales')
plt.show()
````

````python
# Generate a histogram for random data of size 100 using Seaborn:

import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# Generating random data
random_data = np.random.randn(100)

# Creating the histogram
sns.histplot(random_data, bins=10, kde=True, color='purple')
# In the context of Seaborn's histplot function, the kde parameter stands for Kernel Density Estimate.
# When you set kde=True, it adds a smooth curve over the histogram, which represents the estimated probability density function of the data.

# This can be particularly useful for visualizing the distribution of data in a more continuous manner,
# as opposed to the discrete bins of a histogram. The KDE curve helps to identify
# patterns and trends in the data that might not be immediately apparent from the histogram alone.
plt.title('Histogram of Random Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
````

````python
# Create a scatter plot with a regression line using Seaborn:

import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# Generating sample data
np.random.seed(42)
x = np.random.rand(50) * 10  # Random x values
y = 2.5 * x + np.random.randn(50) * 5  # Linear relation with noise

# Creating the scatter plot with regression line
sns.regplot(x=x, y=y, scatter_kws={'color': 'blue'}, line_kws={'color': 'red'})
plt.title('Scatter Plot with Regression Line')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.show()
````
