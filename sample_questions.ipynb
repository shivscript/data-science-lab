**Question 2 (8 Marks)**

**Dataset**  
| Customer_ID | Age | Gender | Transaction_Amount | Product_Category | Payment_Method |  
|------------|----|--------|------------------|----------------|---------------|  
| 101        | 25 | Male   | 500              | Electronics     | Credit Card   |  
| 102        | 34 | Female | 1200             | Clothing        | Debit Card    |  
| 103        | NaN | Male   | 300              | Grocery        | Cash          |  
| 104        | 29 | Female | NaN              | Electronics     | Credit Card   |  

**Instructions:**  

**Step 1: Data Wrangling**  
- Load the dataset into a Pandas DataFrame and display the first five rows.  
- Identify and handle missing values appropriately (e.g., imputation or removal).  
- Encode the categorical variables **Gender**, **Product_Category**, and **Payment_Method** using an appropriate encoding technique.  

**Step 2: Exploratory Data Analysis (EDA)**  
- Generate a summary of numerical and categorical features, including measures like mean, median, and count.  
- Create a **box plot** to visualize the distribution of **Transaction_Amount**.  
- Generate a **bar chart** showing the total transaction amount for each **Product_Category**.  

**Expected Outcome:**  
- A cleaned dataset with missing values handled and categorical variables encoded.  
- Visualizations and statistics explaining key trends in customer transactions.

**Question 3 (8 Marks)**     
You are given a small dataset containing customer information. Perform feature extraction and feature selection to prepare the data for machine learning.

**Dataset**  
| Customer_ID | Age | Gender | Purchase_Frequency | Last_Purchase_Days | Total_Spent | Membership_Level |  
|------------|----|--------|------------------|------------------|------------|----------------|  
| 101        | 25 | Male   | 5                | 10               | 500        | Silver         |  
| 102        | 34 | Female | 2                | 30               | 1200       | Gold           |  
| 103        | 40 | Male   | 7                | 5                | 300        | Silver         |  
| 104        | 29 | Female | 3                | 20               | 700        | Platinum       |  

**Instructions:**  

**Step 1: Feature Extraction**  
- Convert the categorical variables **Gender** and **Membership_Level** into numerical format using an appropriate encoding method.  
- Create a new feature **Customer_Loyalty_Score**, defined as:  

  ***Customer_Loyalty_Score = (Purchase_Frequency / (Last_Purchase_Days + 1)) * Total_Spent***

- Display the updated dataset after adding the new feature.  

**Step 2: Feature Selection**  
- Identify and select the top three most important features using a feature selection technique (e.g., correlation matrix, mutual information, or feature importance from a model).  
- Provide a justification for your selection.  

**Expected Outcome:**  
- A transformed dataset with extracted features and encoded categorical variables.  
- A list of the top three most important features with an explanation of their significance.
